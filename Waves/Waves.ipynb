## Progressive and stationary waves

<a id='Key_terms'></a>
### Key terms

#### Amplitude: 
- maximum displacement of a vibrating particle
- distance from equilibrium to crest of trough

#### Frequency (hertz Hz):
- number of cycles of vibrating particle in a second
- number of complete waves passing a point in a second
- $f = \frac{1}{T}$ where T = time for 1 wave

#### Wavelength ($\lambda$):
- distance between 2 particles at same displacement and velocity at same time
- distance between the same point on the wave cycle to the next/previous

#### Wave speed:
- speed of the wave
- $\frac{\lambda}{f^{-1}} = \lambda f$
- $c = \lambda f$

#### Phase:
- fraction of cycle completed since start of cycle
- phase difference if the distance between 2 particles of same frequency
- phase difference in fraction of cycle = $\frac{d}{\lambda}$
- phase difference in radians = $\frac{2\pi d}{\lambda}$
- phase difference in degrees = $\frac{360 d}{\lambda}$

<a id='Longitudinal_and_transverse_waves'></a>
### Longitudinal and transverse waves

#### Longitudinal waves
- direction of vibration of particles is parallel to direction of wave
- eg. sound, primary seismic, compression waves
- compressions and rarefaction

#### Transverse waves
- direction of vibration of particles is perpendicular to direction of wave
- eg. electromagnetic, secondary seismic, waves on string

<a id='Polarisation'></a>
### Polarisation

- Transverse waves are plane-polarised, vibrations stay in one plane
- if vibrations change plane, its unpolarised
- longitudinal waves cant be polarised
- electromagnetic waves plane is the plane in which the electric field oscillates
- the transmitter and receiver aerials need to be aligned to receive best reception

<a id='Stationary_waves'></a>
### Stationary waves

- When 2 progressive waves of same f meet
- combine to create stationary waves
- nodes: points where particles dont vibrate
- antinodes: points where particles vibrate the most (midpoints between nodes)
- distance between nodes = $\frac{\lambda}{2}$
- first harmonic (fundamental mode of vibration) is when 2 nodes are on each end and 1 antinode in the middle

<img src="img/stationary_waves.png" alt="stationary_waves" style="width: 300px;"/>

for first harmonic:
$f = \frac{1}{2l}\sqrt{\frac{T}{\mu}}$

where $\frac{1}{2l}$ = frequency

and T = tension

and $\mu$ = mass per unit length

<a id='Principle_of_superposition'></a>
### Principle of superposition

##### Superposition: At the point where waves meet, they combine until they move away

##### total displacement = sum of individual displacements

- supercrest: where 2 crests meet and combine
- supertrough: where 2 troughs meet and combine
- can cancel each other out eg. crest and trough
- if amplitude is different its a minimum when canceling each other out

```python
x = np.linspace(0, 4*np.pi, 40*np.pi).astype(np.float32)

y = np.sin(x)
y2 = np.sin(x+np.pi)
y3 = y+y
y4 = y+y2

y5 = 1.2*y2+y


fig, ax = plt.subplots(figsize=(10,5))

ax.plot(x,y, '-g', label='progressive wave')
ax.plot(x,y2, '-r', label='progressive wave in other direction')
ax.plot(x,y3, '--g', label='progressive wave combining')
ax.plot(x,y4, '--r', label='progressive wave canceling')

ax.plot(x, y5, '--b', label='progressive wave minimum')

ax.grid(True)
ax.legend()
plt.show()
```

## Refraction, diffraction and interference

<a id='Interference'></a>
### Interference

#### Coherence

Coherent sources emit waves of same frequency and constant phase difference

eg:
- lasers
- light from source passing through slit and then through 2 different slits

#### Lasers

- Sources of monochromatic light
- coherent

#### Double slit experiment

<img src="img/interference.png" alt="interference" style="width: 300px;"/>

- light from 2 coherent sources of light (2 slits)
- either reinforce or cancels out each other on the screen based on phase and distances
- where reinforced, bright fringe
- where canceled out, dark fringe
- creates an interference pattern
    - when slits too wide, waves from slit interfere with adjacent waves
    - dark fringes get thinner, bright get thicker and contrast is lost


$w = \frac{\lambda D}{s}$

Where:
- s = distance between slit b and c
- D = distance between S2 and F
- w = distance between bright fringe centers

#### With white light

- the spacing depends on the wave length
- the central fringe is white because all wavelengths contribute to this fringe
- the different colours do not overlap exactly because they are spaced out differently

<a id='Diffraction'></a>
### Diffraction

Diffraction is the spreading out of waves after passing through a gap or by an edge

Diffraction increases if:
- gap is narrower
- wavelength is larger

The exiting wave intensity is:
- greatest at centre fringe
- centre fringe is twice as wide as outer fringes
- fringes intensity decreases as further away from centre
- $w = \frac{2\lambda D}{s}$ for centre fringe
- $w = \frac{\lambda D}{s}$ for outer fringes

The fringes are caused by interference from adjacent parts of the slit

#### Diffraction grating

- many closely spaced parallel slits
- parallel beam of monochromatic light directed normally at grating
- light is transmitted in certain directions because light is diffracted and interference

Angle of diffraction
- the 0 order beam is at same angle as the incident
- beams are numbered out from the 0 order beam
- the beams are angled more when:
    - there is a longer wavelength
    - grating has closer slits
- the equation is $d \sin\theta = n\lambda$
- where n is the number order beam
- to find maximum, substitute $\theta = \frac{1}{2}\pi rad$

<a id='Refraction'></a>
### Refraction

Light Ray bends:
- towards the normal when traveling to more dense substance
- away from the normal when traveling to less dense substance

#### Refractive index

The refractive index for a substance s is:

$n_s = \frac{c}{c_s}$

or:

$n = \frac{\sin i}{\sin r}$

<img src="img/refraction.png" alt="refraction" style="width: 300px;"/>

$BQ = ct$
$AP = c_st$
- distance wavefront has traveled
- t = time
- $c$ and $c_s$ = speed of wave out of substance and in respectively
- $\theta_1 = i$ and $\theta_2 = r$

$\sin \theta_1 = \frac{BQ}{AQ}$

$\sin \theta_1 = \frac{ct}{AQ}$

$AQ \sin \theta_1 = ct$

- we use this with the equivalent but for the other part of the wave front

$\sin \theta_2 = \frac{AP}{AQ}$

$\sin \theta_2 = \frac{c_st}{AQ}$

$AQ \sin \theta_2 = c_st$

- we now combine the equalities

$\frac{\sin \theta_1}{\sin \theta_2} = \frac{c}{c_s}$

$n_s = \frac{c}{c_s}$

#### Refraction at a boundary

$\frac{\sin i}{\sin r} = \frac{c_1}{c_2}$

- rewritten as:

$\frac{1}{c_1}\sin i = \frac{1}{c_2}\sin r$

- then multiplied both by $c$ (speed of light in vacuum)

$\frac{c}{c_1}\sin i = \frac{c}{c_2}\sin r$

- now replace the fractions with $n$, the refraction index

$n_1\sin i = n_2\sin r$

#### Refractive index in air

$n \approx 1$
- pretty much the same as in a vacuum

#### White light

$c$ depends upon wavelength in glass so white light gets split into separate wavelengths as their refractive index is different

#### Total internal refraction

<img src="img/refraction-2.png" alt="refraction" style="width: 500px;"/>

- if i = critical angle, the light ray is refracted along the boundary
- if i > critical angle, the light ray undergoes total internal reflection
- $n_1 > n_2$

When i = critical angle

$n_1\sin \theta_c = n_2\sin(\frac{1}{2}\pi rad)$

$n_1\sin \theta_c = n_2$

$\sin \theta_c = \frac{n_2}{n_1}$

#### Optical fibres

<img src="img/optical-fibre.png" alt="refraction" style="width: 300px;"/>

- optical fibre uses total internal reflection so the light rays stay within the core
- the cladding has lower refractive index so light loss is limited
- the cladding also stops light crossing fibres
- the core is very narrow to minimize signal length to prevent modal dispersion
- when rays reflected take longer than rays traveling down centre which could cause signal to overlap with next
- white light causes material dispersion which is when different wavelengths take longer than others
